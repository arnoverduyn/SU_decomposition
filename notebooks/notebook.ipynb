## SU Decomposition Example

### Introduction
In this notebook, the SU decomposition is applied to trajectory data of a real recording of a human-demonstrated rigid body motion. The temporal profile of the U matrix is computed and visualized. 

You can change the value for the regularization parameter L in this notebook. Based on your specified value for L, the notebook provides a comparison of the unregularized U matrix versus the regularized U matrix.

### Running the code

##### 1. Import python libraries and add the data folder to the python path

```python
import sys, pathlib
parent_folder = str(pathlib.Path().resolve().parent)
sys.path.append(parent_folder) 
import src.backend_notebook
path_to_data = rf'{parent_folder}/Data'
path_to_figures = rf'{parent_folder}/figures'
```

##### 2. Specify the type of input trajectory (*only 'pouring' is supported at this time*)

```python
input_trajectory = 'pouring'  
```

##### 3. Load and visualize the trajectory data

```python
T, dt = src.backend_notebook.load_trajectory(input_trajectory,path_to_data)
src.backend_notebook.plot_trajectory(T,path_to_data,path_to_figures)
```

##### 4. Specify a value for the regularization parameter L (expressed in meters)

```python
L = 0.3
```

##### 5. Calculate U from the SU decomposition

```python
U = src.backend_notebook.compute_SU(T,dt,L)
```

##### 6. Visualize the computed U matrix (blue = unregularized, red = regularized)

```python
src.backend_notebook.plot_U(U,dt,path_to_figures)
```
